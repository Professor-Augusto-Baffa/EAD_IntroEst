# Medidas de Dispersão

### Amplitude Total (A)

```python
range_lifeExp = max(data['lifeExp']) - min(data['lifeExp'])
range_year = max(data['year']) - min(data['year'])
print(range_lifeExp)
print(range_year)
```

### Intervalo-Interquartil (d)

```python
import numpy as np
q1_x = np.percentile(data['lifeExp'], 25, interpolation='midpoint')
q3_x = np.percentile(data['lifeExp'], 75, interpolation='midpoint')
print(q3_x - q1_x)
```

```python
from scipy.stats import iqr

print(iqr(data['lifeExp'], rng=(25,75), interpolation='midpoint'))
```

### Desvio Médio (ou desvio absoluto)

### Desvio Padrão

```python
import statistics as st
print(st.pstdev(data['lifeExp']))

import numpy as np
print(np.std(data['lifeExp']))
```

```python
print(data['lifeExp'].std())

import statistics as st
print(st.stdev(data['lifeExp']))
```

### Variância

```python
import statistics as st
print(st.pvariance(data['lifeExp']))

import numpy as np
print(np.var(data['lifeExp']))
```

```python
print(data['lifeExp'].var())

import statistics as st
print(st.variance(data['lifeExp']))
```

### Coeficiente de Variação (CV)

```python
data_mean = data['lifeExp'].mean()

import statistics as st
print(st.pstdev(data['lifeExp'])/data_mean)

import numpy as np
print(np.std(data['lifeExp'])/data_mean)
```

```python
data_mean = data['lifeExp'].mean()

print(data['lifeExp'].std()/data_mean)

import statistics as st
print(st.stdev(data['lifeExp'])/data_mean)
```

### Z-Score

```python
from scipy.stats import zscore
print(zscore(data['lifeExp']))
```

```python
X = 40
Z = (X - data['lifeExp'].mean())/data['lifeExp'].std()
print(data['lifeExp'].mean())
print(Z)
```

# Análise Exploratória de Dados

### Resumo dos Cinco Números

```python
from numpy import percentile

quartis = percentile(data['lifeExp'], [25, 50, 75])
data_min, data_max = data['lifeExp'].min(), data['lifeExp'].max()

print('Min: %.3f' % data_min)
print('Q1: %.3f' % quartis[0])
print('Mediana: %.3f' % quartis[1])
print('Q3: %.3f' % quartis[2])
print('Max: %.3f' % data_max)
```

### Diagrama de Ramos e Folhas

```python
#!pip install stemgraphic
import stemgraphic

df = data[data.year == '2007']
df_africa = data[data.continent == 'Africa']
df_europa = data[data.continent == 'Europe']

stemgraphic.stem_graphic(df_africa['lifeExp'])
stemgraphic.stem_graphic(df_europa['lifeExp'])
```

### Identificação de Valores Discrepantes

```python
from numpy import percentile

quartis = percentile(data['lifeExp'], [25, 75])
d = quartis[1] - quartis[0]
media = data['lifeExp'].mean()
LS = media + 1.5 * d
LI = media - 1.5 * d

print("Limite Superior: %.3f" % LS)
print("Limite Inferior: %.3f" % LI)
```

### Gráfico de Pizza

```python
import matplotlib.pyplot as plt

data_2007 = data[data['year']==2007]
data_2007 = pd.pivot_table(data_2007, values='pop', 
                     index=['continent'], 
                     aggfunc='sum')

plt.pie(data_2007['pop'], labels=data_2007.index, shadow=True, 
        autopct='%1.0f%%', startangle=90)

plt.title('Percentual de habitantes por continente')
```
